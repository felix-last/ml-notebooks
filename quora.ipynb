# Load training data

```python
training_set = pd.read_csv("/Users/gdouzas/Main/data/quora/train.csv")[0:5000]
```

# Text normalization

```python
for question in ["question1", "question2"]:
    training_set[question + "_normalized"] = training_set[question].apply(normalize_text) 
```

```python
training_set["text_diff"] = training_set[["question1_normalized", "question2_normalized"]].apply(lambda texts: text_difference(texts[0], texts[1]), axis=1)
```
